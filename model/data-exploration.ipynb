## Distributions

In this step, I will visualize each distribution of the features.

### Age

```python
plt.figure(figsize=(8, 10), dpi=100)
plt.title("Distribution of the Poems by their Age Features")
plt.pie([df["age"].value_counts()[0], df["age"].value_counts()[1]], labels=["Renaissance", "Modern"], autopct='%1.1f%%', wedgeprops = {"edgecolor" : "black", "linewidth": 2, "antialiased": True})
plt.legend()
plt.show()
```

As seen from the figure above, the blue part represents the percentage of the renaissance poems in the dataset. The orange part represents the percentage of modern poems.

### Type

```python
plt.figure(figsize=(8, 10), dpi=100)
plt.title("Distribution of the Poems by their Type Features")
plt.pie([df["type"].value_counts()[0], df["type"].value_counts()[1], df["type"].value_counts()[2]], labels=["Love", "Nature", "Mythology & Folklore"], autopct='%1.1f%%', wedgeprops = {"edgecolor" : "black", "linewidth": 2, "antialiased": True})
plt.legend()
plt.show()
```

The blue part represents the percentage of the love poems in the dataset. The orange part represents the percentage of nature poems. The green part represents the percentage of poems about mythology or folklore.
